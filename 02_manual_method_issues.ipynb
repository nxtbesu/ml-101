We can plot the dataset, this time we need more power than seaborn so we use matplotlib.

You don't need to try to understand what this code is doing - matplotlib is very powerful (you can find whole courses dedicated to teaching it).

```python
# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python to give us more control over our plot.
# I'm using Matplotlib because I wasn't able to create a 3D chart with Seaborn.
# We only use very basic Matplotlib functionality in this course so you don't need to worry too much about what it is doing.
# At the end of this course, we provide a list of resources for learning some of the third party tools used in this course.

import matplotlib.pyplot as plt

# in particular we want a 3D plot
from mpl_toolkits.mplot3d import Axes3D

# create the plot
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax) 

# plot our data
ax.scatter(
    iris['Sepal_Length'],
    iris['Sepal_Width'],
    iris['Petal_Length'],
    c=iris['Class'].map({ "Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2 }),
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
ax.set_title("Iris")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")

plt.show()
```

Note that we can only plot three features but our dataset has four features.
